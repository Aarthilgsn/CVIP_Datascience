# Exploratory Data Analysis

Performing Exploratory Data Analysis (EDA) on a terrorism dataset is a crucial step to gain insights into the patterns and characteristics of terrorist activities. EDA helps in understanding the data, identifying trends, and discovering potential factors that contribute to terrorism. Here's a general framework for conducting EDA on a terrorism dataset:

1. **Data Collection and Preparation**:
   - Obtain the terrorism dataset from a reliable source, such as the Global Terrorism Database (GTD).
   - Check for missing data, duplicates, and outliers. Handle them appropriately.
   - Ensure the dataset is properly structured, with clear column names and data types.

2. **Data Description**:
   - Begin by summarizing the dataset to understand its basic characteristics.
   - Calculate summary statistics, such as mean, median, and standard deviation for numerical variables.
   - Explore the distribution of categorical variables.

3. **Data Visualization**:
   - Create various types of plots and visualizations to explore the data, including histograms, bar charts, scatter plots, and heatmaps.
   - Visualize the temporal trends of terrorist activities using time series plots.
   - Use geographic visualizations, such as maps, to identify high-risk regions.

4. **Exploring Attack Characteristics**:
   - Analyze the types of terrorist attacks, their frequency, and the severity of their outcomes.
   - Examine the weapon types and tactics commonly used in attacks.
   - Investigate target types and the impact on civilian populations.

5. **Identify Trends and Patterns**:
   - Conduct time-based analysis to uncover trends over the years.
   - Investigate seasonal or periodic patterns in terrorist activities.
   - Identify regions or countries with a higher incidence of terrorism.

6. **Correlation and Association Analysis**:
   - Explore relationships between different variables, such as the correlation between specific attack methods and target types.
   - Use statistical tests to determine if certain factors are associated with a higher likelihood of terrorist incidents.

7. **Hypothesis Testing**:
   - Formulate hypotheses about potential factors influencing terrorism (e.g., socioeconomic factors, political instability) and conduct hypothesis tests to validate or reject these hypotheses.

8. **Geospatial Analysis**:
   - Utilize geospatial data to analyze the distribution of terrorist incidents on a map.
   - Examine hotspots and clusters of terrorist activity.

9. **Identify High-Risk Regions**:
   - Determine regions or countries with a high incidence of terrorism.
   - Analyze the factors contributing to the higher risk in these areas.

10. **Machine Learning and Predictive Modeling** (Optional):
    - Build predictive models to forecast future terrorist incidents based on historical data.
    - Consider using machine learning algorithms for classification or time series forecasting.

11. **Reporting and Communication**:
    - Summarize the key findings, insights, and patterns in a clear and concise manner.
    - Create data visualizations and charts to support your findings.
    - Provide recommendations for policymakers or relevant stakeholders based on your analysis.

12. **Ethical Considerations**:
    - Ensure that the analysis is conducted ethically, respecting privacy and sensitivity of the subject matter.

Remember that EDA is an iterative process, and you may need to revisit earlier steps as you make new discoveries. The goal is to provide actionable insights and inform decision-making related to counterterrorism efforts or policy measures.

# Data Visualization, Storytelling & Experimenting with charts : Understand the relationships between variables

```python
#Chart - 1 visualization code
plt.figure(figsize=(15,7))
sns.countplot(data=data, x='Year')
plt.title('count of terrorist activities each year')
plt.xticks(rotation=90)
plt.show()
```

# Chart - 2

```python
# chart - 2 visualization code
plt.figure(figsize=(15,7))
sns.countplot(data=data, x='Region')
plt.title('Count of terrorist activities by region')
plt.xticks(rotation=90)
plt.show()
```

# Chart - 3

```python
# Chart - 3 visualization code
plt.figure(figsize=(15,7))
sns.lineplot(data=data, x='Year', y='Killed', estimator='sum')
plt.title('Number of people killed by terror attack')
plt.xticks(rotation=90)
plt.show()
```

# Chart - 4

```python
# Chart - 4 visualization code
plt.figure(figsize=(15,7))
sns.countplot(data=data, x='AttackType')
plt.title('Attack types')
plt.xticks(rotation=90)
plt.show()
```

# Chart - 5

```python
# Chart - 5 visualization code
plt.figure(figsize=(15,7))

# Get the count of each group
group_data = data[data['Group'] != 'Unknown']['Group'].value_counts().head(10)

# Use this count for plotting
sns.barplot(x=group_data.index, y=group_data.values)

plt.title('Top 10 terrorist groups with highest no of attacks')
plt.xticks(rotation=90)
plt.xlabel('Group')
plt.ylabel('Count')
plt.show()
```
